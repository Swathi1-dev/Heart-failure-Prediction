```python
sex_encode=LabelEncoder()
ChestPainType_encode=LabelEncoder()
RestingECG_encode=LabelEncoder()
ExerciseAngina_encode=LabelEncoder()
ST_Slope_encode=LabelEncoder()




```

```python
df["Sex"]=sex_encode.fit_transform(df["Sex"])
df["ChestPainType"]=ChestPainType_encode.fit_transform(df["ChestPainType"])
df["RestingECG"]=RestingECG_encode.fit_transform(df["RestingECG"])
df["ExerciseAngina"]=ExerciseAngina_encode.fit_transform(df["ExerciseAngina"])
df["ST_Slope"]=ST_Slope_encode.fit_transform(df["ST_Slope"])
```

```python
with open("sex_encode.pkl","wb")as f:
     pickle.dump(sex_encode,f)

with open("chestpain.pkl","wb")as f:
     pickle.dump(ChestPainType_encode,f)
     
with open("RestingECG_encode.pkl","wb")as f:
     pickle.dump(RestingECG_encode,f)
     
with open("ExerciseAngina_encode.pkl","wb")as f:
     pickle.dump(ExerciseAngina_encode,f)
     
with open("ST_Slope_encode.pkl","wb")as f:
     pickle.dump(ST_Slope_encode,f)
```

```python
scaler=StandardScaler()
```

```python
x=df.drop("HeartDisease",axis=1)
y=df["HeartDisease"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=2)
```

```python
df_train_scaled=scaler.fit_transform(x_train)
df_train_scaled
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(df_train_scaled,y_train)
```

```python
df_xtest_scaled=scaler.transform(x_test)
```

```python
pred=lr.predict(df_xtest_scaled)
```

```python
pred
```

```python
accuracy_score(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

```python
confusion_matrix(y_test,pred)
```

```python
with open("model.pkl","wb")as f:
    pickle.dump(lr,f)
    
with open("scaler.pkl","wb")as f:
    pickle.dump(scaler,f)
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest=RandomForestClassifier(n_estimators=200,n_jobs=10,max_depth=10)
```

```python
random_forest.fit(x_train,y_train)
```

```python
random_pred=random_forest.predict(x_test)
```

```python
accuracy_score(y_test,random_pred)
```

```python
with open("randomforest_model.pkl","wb")as f:
    pickle.dump(random_forest,f)
```
